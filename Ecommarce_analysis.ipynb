```python
Q1 = df['sales_amount'].quantile(0.25)
Q3 = df['sales_amount'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['sales_amount'] < Q1 - 1.5 * IQR) | (df['sales_amount'] > Q3 + 1.5 * IQR)]
print(outliers)

```

```python
# Skewness
skewness = df['sales_amount'].skew()
print("Sales Amount Skewness:", skewness)


```

```python
df.isnull().sum()
```

```python
mode_value = df['region'].mode()[0]
df['region'] = df['region'].fillna(mode_value)
```

```python
df = df.drop(df.index[1000:])
df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df['sales_amount'] = df['sales_amount'].astype(int)
df['quantity_sold'] = df['quantity_sold'].astype(int)
df['discount_(%)'] = df['discount_(%)'].astype(int)
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
plt.bar(df['region'],df['sales_amount'])
plt.title("region by sales")
plt.xlabel("region")
plt.ylabel("sales amount")
plt.show() 
```

```python
df.columns
```

```python
count_paymentmethod = df['payment_method'].value_counts()
plt.pie(count_paymentmethod,autopct = '%1.0f',labels = count_paymentmethod.index)
plt.title("payment method")
plt.axis('equal')
plt.show()
```

```python
product_counts = df['product_category'].value_counts()
plt.pie(product_counts,autopct = '%1.0f',labels = product_counts.index)
plt.show()
```

```python
salaries_by_region = df.groupby('region')['sales_amount'].agg(sum)
plt.pie(salaries_by_region,autopct = '%1.0f',labels = salaries_by_region.index)
plt.title("region by salaries")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.bar(df['payment_method'],df['sales_amount'])
plt.show()
```

```python
plt.hist(df['sales_amount'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Sales Amount')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.boxplot(x='product_category', y='discount_(%)', data=df)
plt.title('Discounts by Product Category')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(df[['sales_amount','quantity_sold','discount_(%)','customer_rating']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['quantity_sold'], df['sales_amount'], alpha=0.6, color='purple')
plt.title('Sales Amount vs Quantity Sold')
plt.xlabel('Quantity Sold')
plt.ylabel('Sales Amount')
plt.show()
```
